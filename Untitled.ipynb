## BUILD AND EVALUATE TIME SERIES FORECAST

#### Splitting the time series into Training and Test Set

```python
train_len =612
train = data[0:train_len] # First 120 lines
test = data[train_len:] 
```

### Stationary Vs non - Stationary

#### Time Series Plot

```python
data['UNRATE'].plot(figsize=(12,4))
plt.legend(loc='best')
plt.title('UnEmployment Rate Data')
plt.show()
```

### Agumented Dickey-Fuller Test(ADF)

```python
from statsmodels.tsa.stattools import adfuller
adf_test = adfuller(data['UNRATE'])

print('ADF statstic: %f' % adf_test[0])
print('Critical Value @ 0.05: %.2f' % adf_test[4]['5%'])
print('p-value: %f' % adf_test[1])
```

### KPSS Test

```python
from statsmodels.tsa.stattools import kpss
kpss_test = kpss(data['UNRATE'])

print('KPSS statstic: %f' % kpss_test[0])
print('Critical Value @ 0.05: %.2f' % kpss_test[3]['5%'])
print('p-value: %f' % kpss_test[1])
```

## Non Stationary to Stationary

#### BOX COX Transformation

```python
from scipy.stats import boxcox
data_boxcox = pd.Series(boxcox(data['UNRATE'], lmbda=0), index = data.index)
plt.figure(figsize=(12,4))
plt.plot(data_boxcox, label='After Box Cox transformation')
plt.legend(loc='best')
plt.title('After Box Cox transformation')
plt.show()
```

### The trend is still thee so we  do differencing

```python
data_boxcox_diff =pd.Series(data_boxcox - data_boxcox.shift(), index=data.index)
data_boxcox_diff.dropna(inplace=True)

plt.figure(figsize=(12,4))
plt.plot(data_boxcox_diff, label='After Box Cox transform and differencing')
plt.title('After Box Cox transform and differencing')

plt.show()
```

### ADF TEST AFTER BOXCOX TRANSFORMATION

```python
adf_test = adfuller(data_boxcox_diff)

print('ADF statstic: %f' % adf_test[0])
print('Critical Value @ 0.05: %.2f' % adf_test[4]['5%'])
print('p-value: %f' % adf_test[1])
```

### KPSS TEST AFTER BOXCOX TRANSFORAMTION


```python

kpss_test = kpss(data_boxcox_diff)

print('KPSS statstic: %f' % kpss_test[0])
print('Critical Value @ 0.05: %.2f' % kpss_test[3]['5%'])
print('p-value: %f' % kpss_test[1])
```

### Auto Correlation Funtion(ACF) PLOT

```python
from statsmodels.graphics.tsaplots import plot_acf
plt.figure(figsize =(12,4))
plot_acf(data_boxcox_diff, ax=plt.gca(), lags=30)
plt.show()
```

### PArtial Autocorrelation(PACF) Plot

```python
from statsmodels.graphics.tsaplots import plot_pacf
plt.figure(figsize=(12,4))
plot_pacf(data_boxcox_diff, ax=plt.gca(), lags=30)
plt.show()
```

### Splitting Into Training and Test Data Sets

```python
train_data_boxcox = data_boxcox[:train_len]
test_data_boxcox = data_boxcox[train_len:]
train_data_boxcox_diff = data_boxcox_diff[:train_len-1]
test_data_boxcox_diff = data_boxcox_diff[train_len -1:]
```

### Auto Regression Method(AR)

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
model = ARIMA(train_data_boxcox_diff, order=(12,0,0))
```

```python
model_fit = model.fit()
print(model_fit.params)
```

### Recover Original Time Series Forecast

```python
y_hat_ar = data_boxcox_diff.copy()
y_hat_ar['ar_forecast_boxcox_diff'] = model_fit.predict(data_boxcox_diff.index.min(), data_boxcox_diff.index.max())
y_hat_ar['ar_forecast_boxcox'] = y_hat_ar['ar_forecast_boxcox_diff'].cumsum()
y_hat_ar['ar_forecast_boxcox'] = y_hat_ar['ar_forecast_boxcox'].add(data_boxcox[0])
y_hat_ar['ar_forecast'] = np.exp(y_hat_ar['ar_forecast_boxcox'])
```

### PLOT TRAIN, TEST AND FORECAST

```python
plt.figure(figsize=(12,4))
plt.plot(train['UNRATE'], label = 'Train')
plt.plot(test['UNRATE'], label = 'Test')
plt.plot(y_hat_ar['ar_forecast'][test.index.min():], label = 'Auto Regression Forecast')
plt.legend(loc='best')
plt.title('Auto Regression Method')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

rmse = np.sqrt(mean_squared_error(test['UNRATE'], y_hat_ar['ar_forecast'][test.index.min():])).round(2)
mape = np.round(np.mean(np.abs(test['UNRATE']-y_hat_ar['ar_forecast'][test.index.min():])/test['UNRATE'])*100,2)

results = pd.DataFrame({'Method': ['Auto Regressive (AR) method'], 'RMSE':[rmse], 'MAPE':[mape]})
### results = pd.concat([results, tempResults])
results = results[['Method','RMSE','MAPE']]
results
```

```python
print('Accuracy', (1-(mape/100)) * 100)
```

### Calculate RMSE AND MAPE

```python
# Calculate the start and end indices for the test data
start_index = test.index.min()
end_index = test.index.max()

# Ensure the forecasted data is aligned with the test data
# Note: If the index is datetime, use .loc to properly align the slices
forecasted_values = y_hat_arima['arima_forecast'].loc[start_index:end_index]

# Now calculate RMSE and MAPE using the correctly aligned data
rmse = np.sqrt(mean_squared_error(test['UNRATE'], forecasted_values)).round(2)
mape = np.round(np.mean(np.abs(test['UNRATE'] - forecasted_values) / test['UNRATE']) * 100, 2)

# Rest of the code to store the results
tempResults = pd.DataFrame({'Method': ['ARIMA method'], 'RMSE': [rmse], 'MAPE': [mape]})
results = pd.concat([results, tempResults])
results = results[['Method', 'RMSE', 'MAPE']]
print(results)
```
