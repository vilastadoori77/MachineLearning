### Outlier Detection

#### Box Plot and Interquartile Range

```python
import seaborn as sns
fig = plt.subplots(figsize=(12,2))
ax = sns.boxplot(x = data['Temp'], whis=1.5)
```

```python
# Calculate the first quartile (Q1)
Q1 = data['Temp'].quantile(0.25)

# Calculate the third quartile (Q3)
Q3 = data['Temp'].quantile(0.75)

# Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1

# Print the IQR
print(f"The interquartile range (IQR) is: {IQR}")
```

```python
Outlier1 = Q3 + IQR*1.5
print(f"Outlier1  is: {Outlier1}")
```

```python
# However, the right whisker does not extend past the maximum data point
# So, we need to determine the maximum value within the right whisker's range
right_whisker_endpoint = data['Temp'][data['Temp'] <= Outlier1].max()

# Print the right whisker endpoint
print(f"The right whisker extends to: {right_whisker_endpoint}")
```

```python
Outlier2 = Q1 - IQR*1.5
print(f"Outlier1  is: {Outlier2}")
```

```python
# However, the right whisker does not extend past the maximum data point
# So, we need to determine the maximum value within the right whisker's range
left_whisker_endpoint = data['Temp'][data['Temp'] <= Outlier2].max()

# Print the right whisker endpoint
print(f"The left whisker extends to: {left_whisker_endpoint}")
```

## BUILD AND EVALUATE TIME SERIES FORECAST

## Splitting the time series into Training and Test Set


```python
train_len = 511
train = data[0:train_len] # First 584 lines
test = data[train_len:] 
```

### Naive method


```python
y_hat_naive = test.copy()
###Calculating the next value based on the navive method
y_hat_naive['naive_forecast'] = train['Temp'][train_len -1]
```

### plot , test and forecast

```python
plt.figure(figsize=(12,4))
plt.plot(train['Temp'], label = 'Train')
plt.plot(test['Temp'], label = 'Test')
plt.plot(y_hat_naive['naive_forecast'], label='Navie forecast')
plt.legend(loc='best')
plt.title('Naive Method')
plt.show()
```

```python
# Assuming 'data' is a pandas DataFrame with a time series column named 'Sales'
last_observed_value = data['Temp'].iloc[-1]

# Create a naive forecast for the next 'n' periods
n_periods = 12  # For example, forecast for the next 12 months
naive_forecast = [last_observed_value] * n_periods

# naive_forecast now contains the same value repeated for n_periods times
```

```python
print(naive_forecast)
```

### Simple Average Method

```python
y_hat_avg = test.copy()
### Calculate the next value based on the simple average method.

y_hat_avg['avg_forecast'] = train['Temp'].mean()
```

### Plot Train , Test and Forecast

```python
plt.figure(figsize=(12,4))
plt.plot(train['Temp'], label = 'Train')
plt.plot(test['Temp'], label = 'Test')
plt.plot(y_hat_avg['avg_forecast'], label='Simple Avg forecast')
plt.legend(loc='best')
plt.title("Simple Avg Method")
plt.show()
```

```python
# Assuming 'data' is a pandas DataFrame with a time series column named 'Sales'

# Calculate the simple average
simple_average = data['Temp'].mean()

# Create a simple average forecast for the next 'n' periods
n_periods = 12  # For example, forecast for the next 12 months
simple_average_forecast = [simple_average] * n_periods

# simple_average_forecast now contains the simple average value repeated for n_periods times
```

```python
print(simple_average_forecast)
```

### Calculate the RMSE and MAPE to evaluate forecast for the Naive Method


```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(test['Temp'], y_hat_naive['naive_forecast'])).round(2)
mape = np.round(np.mean(np.abs(test['Temp']-y_hat_naive['naive_forecast'])/test['Temp'])*100,2)
results = pd.DataFrame({'Method' :['Naive method'], 'MAPE':[mape], 'RMSE':[rmse]})
results = results[['Method', 'RMSE', 'MAPE']]
results
```

```python
#Accuracy for Naive Average Method

accuracy = (1-(mape/100))*100
print("Accuracy", accuracy)
```

### Calculate RMSE and MAPE to evaluate forecast


```python
rmse = np.sqrt(mean_squared_error(test['Temp'], y_hat_avg['avg_forecast'])).round(2)
mape = np.round(np.mean(np.abs(test['Temp']-y_hat_avg['avg_forecast'])/test['Temp'])*100,2)
tempResults = pd.DataFrame({'Method':['Simple avg method'], 'RMSE':[rmse], 'MAPE':[mape]})
results = pd.concat([results, tempResults])
results
```

```python
#Accuracy for Simple Average Method

accuracy = (1-(mape/100))*100
print("Accuracy", accuracy)
```

### Simple Moving Average

```python
y_hat_sma = data.copy()
ma_window = 3
y_hat_sma['sma_forecast'] = data['Temp'].rolling(ma_window).mean() #Rolling 12 month average
y_hat_sma['sma_forecast'][train_len:] = y_hat_sma['sma_forecast'][train_len-1]
```

### plot train, test and forecast

```python
plt.figure(figsize=(12,4))
plt.plot(train['Temp'], label = 'Train')
plt.plot(test['Temp'], label = 'Test')
plt.plot(y_hat_sma['sma_forecast'], label = 'Simple moving average forecast')
plt.legend(loc='best')
plt.title('Simple Moving Average Method')
plt.show()
```

### Calculate RMSE and MAPE to evaluate forecast of Simple Moving Average.

```python
rmse = np.sqrt(mean_squared_error(test['Temp'], y_hat_sma['sma_forecast'][train_len:])).round(2)
mape = np.round(np.mean(np.abs(test['Temp']-y_hat_sma['sma_forecast'][train_len:])/test['Temp'])*100,2)

tempResults = pd.DataFrame({'Method': ['Simple Moving Average Method'], 'RMSE':[rmse], 'MAPE':[mape]})
results = pd.concat([results, tempResults])
results
```

```python
#Accuracy for Simple Average Method

accuracy = (1-(mape/100))*100
print("Accuracy", accuracy)
```

### Simple Exponential Smoothing

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
model = SimpleExpSmoothing(train['Temp'])
model_fit = model.fit(smoothing_level=0.2, optimized=False)
model_fit.params
y_hat_ses = test.copy()
y_hat_ses['ses_forecast'] = model_fit.forecast(len(test['Temp']))
```

### Plot Train, Test and forecast

```python
plt.figure(figsize=(12,4))
plt.plot(train['Temp'], label='Train')
plt.plot(test['Temp'], label='Test')
plt.plot(y_hat_ses['ses_forecast'], label='Simple Exponential Smooting Forecast only Level and no Trend or Seasonality')
plt.legend(loc='best')
plt.title('Simple Exponential Smoothing Method')
plt.show()
```

```python
print(test['Temp'].isnull().sum()) # Check for Nan's in test data
print(y_hat_ses['ses_forecast'].isnull().sum())  # Check for NaNs in forecast data
```

### Calculate RMSE and MAPE

```python
rmse = np.sqrt(mean_squared_error(test['Temp'], y_hat_ses['ses_forecast'])).round(2)
mape = np.round(np.mean(np.abs(test['Temp']-y_hat_ses['ses_forecast'])/test['Temp'])*100,2)

tempResults = pd.DataFrame({'Method': ['Simple Exponential Smoothing Method'], 'RMSE':[rmse], 'MAPE':[mape]})
results = pd.concat([results, tempResults])
results = results[['Method','RMSE','MAPE']]
results
```

```python
accuracy = (1-(mape/100))*100
print("Accuracy", accuracy)
```

### Holts Exponential Smoothing Method with Trend

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
model = ExponentialSmoothing(np.asarray(train['Temp']), seasonal_periods=3, trend='additive', seasonal=None)
model_fit = model.fit(smoothing_level=0.5,smoothing_slope=0.01,optimized=False)
model_fit.params
print(model_fit.params)
y_hat_holt = test.copy()
y_hat_holt['holt_forecast'] = model_fit.forecast(len(test['Temp']))
```

```python
plt.figure(figsize=(12,4))
plt.plot(train['Temp'], label='Train')
plt.plot(test['Temp'], label='Test')
plt.plot(y_hat_holt['holt_forecast'], label='Holt\'s exponential Smoothing forecast')
plt.legend(loc='best')
plt.title('Holt\s exponential smoothing forecast')
plt.show()
```

```python
rmse = np.sqrt(mean_squared_error(test['Temp'], y_hat_holt['holt_forecast'])).round(2)
mape = np.round(np.mean(np.abs(test['Temp']-y_hat_holt['holt_forecast'])/test['Temp'])*100,2)

tempResults = pd.DataFrame({'Method': ['Holt\'s Exponential Smoothing Method'], 'RMSE':[rmse], 'MAPE':[mape]})
results = pd.concat([results, tempResults])
results = results[['Method','RMSE','MAPE']]
results
```
